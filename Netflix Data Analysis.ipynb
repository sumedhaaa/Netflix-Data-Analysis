```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('netflix_titles.csv')
dataset.head()
```

```python
movie_df=dataset[dataset['type']=='Movie']
movie_df.head()
release_year=movie_df.groupby('release_year').size().reset_index().rename(columns={0:'Total Movies'})
release_year.sort_values(['Total Movies'], ascending=False)
release_year.to_excel('release_year.xlsx')
```

```python
dataset['new_date']=pd.to_datetime(dataset['date_added'])
dataset['year_added']=dataset['new_date'].dt.year
year_added=dataset.groupby('year_added').size().reset_index().rename(columns={0:'Total'})
year_added.sort_values('Total',ascending=False)
year_added.to_excel('year_added.xlsx')
```

```python
dataset['month_added']=dataset['new_date'].dt.month
month_added=dataset.groupby('month_added').size().reset_index().rename(columns={0:'Total'})
month_added.sort_values('Total',ascending=False)
month_added.to_excel('month_added.xlsx')
```

```python
movie_df['title_len']=[len(title) for title in movie_df['title']]
movie_df.head()
```

```python
top_title_lenghts=movie_df.sort_values('title_len',ascending=False)
top_title_lenghts
```

```python
title_and_cast=dataset[['title','cast']]
title_and_cast=title_and_cast.dropna()
cast_names_counter={}
for index, row in title_and_cast.iterrows():
    movie_cast=row['cast']
    movie_cast_split=movie_cast.split(',')
    movie_cast_stripped=[name.strip().lower() for name in movie_cast_split]
    for name in movie_cast_stripped:
        cast_names_counter[name]=cast_names_counter.get(name,0)+1
    
dict(sorted(cast_names_counter.items(), key=lambda item: item[1], reverse=True))


    
```

```python
cast_df=pd.DataFrame.from_dict(dict(sorted(cast_names_counter.items(), key=lambda item: item[1], reverse=True)),orient='index')
cast_df
cast_df.to_excel('cast_df.xlsx')
```

```python
title_and_country=dataset[['title','country']]
title_and_country=title_and_country.dropna()
```

```python
country_counter={}
for index, row in title_and_country.iterrows():
    movie_country=row['country']
    movie_country_split=movie_country.split(',')
    movie_country_stripped=[name.strip().lower() for name in movie_country_split]
    for name in movie_country_stripped:
        country_counter[name]=country_counter.get(name,0)+1
    
dict(sorted(country_counter.items(), key=lambda item: item[1], reverse=True))
```

```python
country_df=pd.DataFrame.from_dict(dict(sorted(country_counter.items(), key=lambda item: item[1], reverse=True)),orient='index')
country_df.to_excel('country_df.xlsx')
```
